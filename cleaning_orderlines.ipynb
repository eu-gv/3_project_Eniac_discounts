## fix unit_price in order_lines (from total_paid of orders)

```python
import pandas as pd

ord_l = pd.read_csv('orderlines.csv')
ords = pd.read_csv('orders.csv')
```

## orderlines without any id_order duplicates, excluding also first occurrence, and copy

```python
ord_l_df.info()
print('\n', ord_l_df.drop_duplicates(subset=['id_order'], keep=False)['id_order'].count(), 'rows of duplicates')
print(ord_l_df.shape[0], 'total rows')
print(ord_l_df.shape[0] - ord_l_df.drop_duplicates(subset=['id_order'], keep=False)['id_order'].count(), 'discarded rows')
#ord_l_df.loc[~ord_l_df.id_order.duplicated()]
```

```python
# new df without duplicates:
ord_l_df_no_dupl = ord_l_df.drop_duplicates(subset=['id_order'], keep=False)
ord_l_df_no_dupl.nunique()
ord_l_df_no_dupl.info()
```

```python
# count number of corrupted prices in new df without duplicates:
ord_l_df_no_dupl.loc[ord_l_df_no_dupl.unit_price.str.contains("\d+\.\d+\.\d+"), 'unit_price'].count()
```

```python
del ord_l_df_no_dupl['product_id']
```

```python
# sort ordersL with corrupted prices by the product_quantity:
ord_l_df_no_dupl.loc[ord_l_df_no_dupl.unit_price.str.contains("\d+\.\d+\.\d+")].sort_values(by='product_quantity')
```

```python
# filter ordersL with corrupted prices & product_quantity==1:
ord_l_df_no_dupl.loc[(
            (ord_l_df_no_dupl.unit_price.str.contains("\d+\.\d+\.\d+")) &
            (ord_l_df_no_dupl['product_quantity']==1) ),
            'id_order'
                    ]
```

```python
# list of id orders: ONLY from ordersL with corrupted prices that also have product_quantity==1:
corrupted_ids = ord_l_df_no_dupl.loc[(
                            (ord_l_df_no_dupl.unit_price.str.contains("\d+\.\d+\.\d+")) &
                            (ord_l_df_no_dupl['product_quantity']==1) ),
                            'id_order'
                                    ]
```

## copy of orders

```python
# copy of orders filtered by the ids with ONLY corrupted prices from orderL
ords_df_cp = ords_df[ords_df['order_id'].isin(corrupted_ids)].copy()
ords_df_cp
```
